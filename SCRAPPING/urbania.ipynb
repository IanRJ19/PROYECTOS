```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from decouple import config
import time
import pandas as pd 

# Inicializar el driver de Selenium
driver = webdriver.Chrome()

# Ir a la página de inicio de LinkedIn
driver.get("https://urbania.pe/buscar/alquiler-de-departamentos-en-pueblo-libre--lima--lima?bedroomMin=3&coveredAreaMin=80&priceMax=2600&currencyId=6")


time.sleep(9)

```

```python

job_elements = driver.find_elements(By.CSS_SELECTOR,"[data-qa='posting PROPERTY']")

jobs = []
```

```python
job_elements
```

```python
for element in job_elements:
    print(element.text)
```

```python
time.sleep(9)
# Iterar sobre los elementos de las ofertas de trabajo
for job_element in job_elements:
    Precio_elements = job_element.find_elements(By.CSS_SELECTOR,"[data-qa='POSTING_CARD_PRICE']")
    #company_elements = job_element.find_elements(By.CSS_SELECTOR,"[data-qa='POSTING_CARD_GALLERY']")
    Ubicacion_elements = job_element.find_elements(By.CSS_SELECTOR,"[data-qa='POSTING_CARD_LOCATION']")
    featueres_elements = job_element.find_elements(By.CSS_SELECTOR,"[data-qa='POSTING_CARD_FEATURES']")
    Descripcion_elements = job_element.find_elements(By.CSS_SELECTOR,"[data-qa='POSTING_CARD_DESCRIPTION']")
    
    Precio = Precio_elements[0].text if Precio_elements else None
    #company = company_elements[0].text if company_elements else None
    Ubicacion = Ubicacion_elements[0].text if Ubicacion_elements else None
    features = featueres_elements[0].text if featueres_elements else None
    Descripcion=Descripcion_elements[0].text if Descripcion_elements else None
    jobs.append({"Precio": Precio, "Ubicacion": Ubicacion,"caracteristicas":features,"Descripcion":Descripcion})


# Almacenar los datos en un DataFrame de Pandas
jobs_df = pd.DataFrame(jobs)
```

```python
jobs_df.columns
jobs_df["Ubicacion"]=jobs_df["Ubicacion"].str.replace(" ", "")
```

```python
jobs_df["caracteristicas"]
df = jobs_df["caracteristicas"].str.split('\n', expand = True)
```

```python
jobs_df=pd.merge(jobs_df,df,left_index=True, right_index=True)
```

```python
jobs_df=jobs_df.drop("caracteristicas",axis=1)
```

```python

jobs_df=jobs_df.rename(columns={4: 'Estacionamientos',3:"Baños",2:"Dormitorios",0:"M2 Total",1:"M2 Techada"})
```

```python
jobs_df["Ubicacion"]=jobs_df["Ubicacion"].astype(str).str.strip()

```

```python
jobs_df.to_excel("Ban.xlsx")
```
