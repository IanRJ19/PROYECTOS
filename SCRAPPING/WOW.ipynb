```python
import pandas as pd
import numpy as np

datas=[]

for i in range(1, 4):

    from selenium import webdriver
    from selenium.webdriver.common.by import By
    from selenium.webdriver.support.ui import WebDriverWait
    from selenium.webdriver.support import expected_conditions as EC


    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless')
    driver = webdriver.Chrome(options=chrome_options)


    url = f'https://check-pvp.fr/ranking/eu/all-realms/all-factions/all-classes/all-specs/rateatm3v3/desc/{i}'
    driver.get(url)
    print(i)
    table = WebDriverWait(driver, 60).until(EC.presence_of_element_located((By.XPATH, '//table')))
    table_rows = table.find_elements(By.TAG_NAME, 'tr')
    print(f"Se encontraron {len(table_rows)} filas en la tabla de la página {i}")
    data = []

    for row in table_rows:
        row_data = []
        cells = row.find_elements(By.TAG_NAME, 'td')
        for cell in cells:
            row_data.append(cell.text)
        data.append(row_data) 

    df = pd.DataFrame(data)

    import numpy as np
    e=df
    e=e.dropna()
    e = e.replace('', np.nan)
    # Elimina las columnas con valores en vacío (NaN)
    e = e.dropna(axis=1, how='any')
    datas.append(e)

driver.quit()


df = pd.concat(datas)
e=df
e=e.dropna()
e = e.replace('', np.nan)
e = e.dropna(axis=1, how='any')
b=pd.DataFrame()
b['RANKING']=e[0]
b['NOMBRE']=e[1]
b['2c2']=e[2]
b['3c3']=e[9]
b['RBG']=e[16]
b['2c2EXP']=e[23]
b['3c3EXP']=e[24]
b['RBGEXP']=e[25]
b['ARENAPOINT']=e[26]
b['RBGPOINT']=e[27]
b['ILVL']=e[28]
b['PUNTOSDELOGRO']=e[29]
b['VISTAS']=e[30]

b=b.set_index('RANKING')
b
```

Este código extrae datos de una página web y crea un DataFrame de Pandas con los datos extraídos.

Comienza importando las bibliotecas Pandas y Numpy. Luego, crea una lista vacía llamada datas para almacenar los datos extraídos de cada página.

Después, entra en un bucle for que itera de 1 a 3 y dentro de él, importa las bibliotecas Selenium y define algunas opciones para el navegador Chrome. A continuación, crea una URL y abre el navegador para acceder a la página. Luego, utiliza Selenium para encontrar la tabla en la página y extraer sus filas y celdas. Los datos extraídos se almacenan en una lista llamada data.

Después de haber extraído los datos de la tabla, crea un DataFrame de Pandas llamado df con ellos. A continuación, se eliminan las filas con valores vacíos (NaN) y se eliminan las columnas que contienen valores vacíos utilizando la función dropna(). Los datos procesados se almacenan en la variable e, que se agrega a la lista datas.

Una vez que se ha completado el bucle for, se crea un DataFrame de Pandas b que concatena todos los datos almacenados en la lista datas. Luego, se eliminan las filas vacías y se eliminan las columnas con valores vacíos. Los datos se asignan a las columnas del DataFrame b.

Por último, se establece la columna "RANKING" como el índice del DataFrame b.

En resumen, el código extrae datos de una página web utilizando Selenium, procesa los datos y los almacena en un DataFrame de Pandas. Luego, se realiza una limpieza de datos y se almacenan en un nuevo DataFrame b.
