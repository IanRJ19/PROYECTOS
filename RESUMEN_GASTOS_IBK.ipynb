```python
import os
import pandas as pd
#Lectura de archivos
ruta="C:/Users/Rayzek/Downloads"
archivos=os.listdir(ruta)
df_ESTADOS = []

for i in range(len(archivos)):
    a=archivos[i]
    if ("txt" in a):
        print(a)
        b=pd.read_csv(ruta+"/"+a,skiprows=205,dtype=str,on_bad_lines='skip',encoding='latin-1')
        b["S/ US$"]=b
        b=b["S/ US$"].str.split(" ",4, expand = True)
        b.columns=["1","2","3","4","5"]

        a=b[b["1"] == 'SUBTOTAL'].index[0]
        b=b[0:a]
        df_ESTADOS.append(b)
############################################################################################
#CONSTANCIA 
Constancias=pd.concat(df_ESTADOS)
Constancias.to_excel("C:/Users/Rayzek/Downloads/FINAL.xlsx")
print(Constancias)
```
