```python
from tkinter import *
from tkinter import filedialog
import os
import time

ruta="H:/Mi unidad/TRABAJO/PROCESOS_AUTOMATICOS/Script_CaptaciónACK"

semana_inicio="2022-10-17"
semana_fin="2022-11-04"
fecha_inicio="2022-10-17"
fecha_fin="2022-11-04"

import pandas as pd
import time
import os
import time
start_time = time.time()

#Lectura de archivos
archivos=os.listdir(ruta)
df_reporte = []

for i in range(len(archivos)):
    a=archivos[i]
    if ("ReporteAzteca" in a):
        print(a)
        b=pd.read_excel(ruta+"/"+a,dtype=str,sheet_name="acumulado")
        df_reporte.append(b)
    elif ("Zonas" in a):
        zonas=pd.read_excel(ruta+"/"+a,dtype=str,sheet_name="Inicio Octubre")
        print("No es un reporte "+a)
    else:
        print("No es un reporte "+a)

unido=pd.concat(df_reporte)
unido=unido.drop_duplicates(subset=unido.columns)
unido.to_excel(ruta+"/"+"Reportes_UNIDOS.xlsx")

#INICIAMOS EJECUCIÓN DE FORMATEADO
tabla=unido

#Ordenamos por Fecha y Eliminamos DNI duplicados
tabla["Fecha"]=pd.to_datetime(tabla["Fecha"])


tabla["KEY"]=tabla["Hora"]+tabla["Terminal"]+tabla["Terminal"]+tabla["Id"]+tabla["Distrito"]
tabla=tabla.drop_duplicates(subset="KEY")



#Filtramos la fecha
tabla=tabla[(tabla['Fecha'] >= semana_inicio) & (tabla['Fecha'] <= semana_fin)]
tabla=tabla.sort_values(["Fecha","Hora"], ascending=True)

#a=tabla[tabla.duplicated(subset="DNI")]

a=tabla

tabla=tabla.drop_duplicates(subset="DNI")



#Filtramos la fecha CAPTACIÓN 
tabla=tabla[(tabla['Fecha'] >= fecha_inicio) & (tabla['Fecha'] <= fecha_fin)]
tabla=tabla.sort_values(["Fecha","Hora"], ascending=True)

#Cortamos la fecha
tabla["Fecha"] =tabla["Fecha"].astype("str")
tabla["Fecha"] = tabla["Fecha"].str.slice(start=0, stop=10)

#Creando F1
tabla["F1"]=tabla["Terminal"]+tabla["Fecha"]



#Filtramos la fecha CAPTACIÓN UNICOS
tabla=tabla[(tabla['Fecha'] >= fecha_inicio) & (tabla['Fecha'] <= fecha_fin)]
tabla=tabla.sort_values(["Fecha","Hora"], ascending=True)
#Cortamos la fecha
tabla["Fecha"] =tabla["Fecha"].astype("str")
tabla["Fecha"] = tabla["Fecha"].str.slice(start=0, stop=10)
#Creando F1
tabla["F1"]=tabla["Terminal"]+tabla["Fecha"]

#Filtramos la fecha CAPTACIÓN CON DUPLICADOS
a=a[(a['Fecha'] >= fecha_inicio) & (a['Fecha'] <= fecha_fin)]
a=a.sort_values(["Fecha","Hora"], ascending=True)
#Cortamos la fecha
a["Fecha"] =a["Fecha"].astype("str")
a["Fecha"] = a["Fecha"].str.slice(start=0, stop=10)
#Creando F1
a["F1"]=a["Terminal"]+a["Fecha"]



```
